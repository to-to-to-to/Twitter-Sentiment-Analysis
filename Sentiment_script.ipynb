```python
hasilAnalisis = []


for tweet in hasilSearch:
    tweet_properties = {}
    tweet_properties["Date"] = tweet.created_at
    tweet_properties["UserName"] = tweet.user.screen_name
    tweet_properties["Tweet"] = tweet.text
    tweet_cleaned = ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)"," ",tweet.text).split())
    
    analysis = TextBlob(tweet_cleaned)
    try:
        analysis = analysis.translate(to = "en")
    except Exception as e:
        print(e)
    
    if analysis.sentiment.polarity > 0.0 :
        tweet_properties["sentimen"] = "Positive"
    elif analysis.sentiment.polarity == 0.0 :
        tweet_properties["sentimen"] = "Neutral"
    else:
        tweet_properties["sentimen"] = "Negative"
    
    
    
    if tweet.retweet_count > 0:
        if tweet_properties not in hasilAnalisis:
            hasilAnalisis.append(tweet_properties)
        else:
            hasilAnalisis.append(tweet_properties)
```

```python
tweet_positif = [t for t in hasilAnalisis if  t["sentimen"] == "Positive"]
tweet_netral = [t for t in hasilAnalisis if  t["sentimen"] == "Neutral"]
tweet_negatif = [t for t in hasilAnalisis if  t["sentimen"] == "Negative"]
```
